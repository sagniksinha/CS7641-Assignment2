```python
import numpy as np
import matplotlib.pyplot as plt
import time
import utils
import pandas as pd
import gc
import mlrose_hiive as mlrose

from mlrose_hiive.algorithms.decay import ExpDecay
from mlrose_hiive.neural import NeuralNetwork

from sklearn.metrics import log_loss, classification_report, accuracy_score, f1_score
from sklearn.model_selection import train_test_split

class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    END = '\033[0m'
```

```python
def replace_algorithm(value):
    if value == 1:
        return mlrose.GeomDecay()
    elif value == 2:
        return mlrose.ExpDecay()
    else:
        return value
replace_algorithm(2)
```

```python
def SA_DecayFn(trainX, trainY, testX, testY):
    
    gc.collect()
    max_iters = [100 * (i+1) for i in range(10)]
    seeds = [1,123,300,999,12345]
    schedule_decays = [mlrose.GeomDecay(), mlrose.ExpDecay(), mlrose.ArithDecay()]
    schedule_decays_names = ['Geom Decay', 'Exp Decay', 'Arith Decay']
    color = ['blue', 'green', 'purple']
    algorithm = 'simulated_annealing'
    header = 24
    other_title = 18
    Output = pd.DataFrame(columns=['algorithm', 'random_seed', 'Itterations', 'schedule_decays', 'train_accuracy', 'test_accuracy'])
    
    for max_iter in max_iters:
        for i, schedule_decay in enumerate(schedule_decays):
            for seed in seeds:
                gc.collect()
                # Define neural network classifier
                nn = NeuralNetwork(hidden_nodes=[10,], activation='relu',
                                       algorithm=algorithm, max_iters=int(max_iter),
                                       bias=True, is_classifier=True, learning_rate=0.1,schedule=replace_algorithm(i),
                                       early_stopping=False, random_state=seed, curve=True)

                nn.fit(trainX, trainY)
                      
                train_accuracy = accuracy_score(trainY, pd.DataFrame(nn.predict(trainX), index=trainY.index))
                test_accuracy = accuracy_score(testY, pd.DataFrame(nn.predict(testX), index=testY.index))

                Output = Output.append({'algorithm' : algorithm,
                                        'random_seed' : seed,
                                        'Itterations' : max_iter,
                                        'schedule_decays' : schedule_decays_names[schedule_decays.index(schedule_decay)],
                                        'train_accuracy' : train_accuracy,
                                        'test_accuracy' : test_accuracy}, ignore_index=True)
                display(Output)
    
    Output = Output.groupby(['schedule_decays', 'Itterations']).agg({'train_accuracy': ['mean', 'std'], 'test_accuracy': ['mean', 'std']}).reset_index()
    # Rename the columns
    Output.columns = ['schedule_decays', 'Itterations', 'mean_train_accuracy', 'std_train_accuracy', 'mean_test_accuracy', 'std_test_accuracy']
    
    
    #################################
    ####plot train accuracy
    #################################
    # Plot fitness vs iteration with separate lines for each temperature
    plt.clf()
    plt.figure(figsize=(12, 8))
    plt.gcf().set_facecolor('white')
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.set_facecolor('white')
    ax.set_facecolor('white')
    for schedule_decay in schedule_decays:
        print(color[schedule_decays.index(schedule_decay)])
        temp_data = Output[Output['schedule_decays'] == schedule_decays_names[schedule_decays.index(schedule_decay)]]
        plt.plot(temp_data['Itterations'], temp_data['mean_train_accuracy'], label=f'{schedule_decays_names[schedule_decays.index(schedule_decay)]} (Train)', color=color[schedule_decays.index(schedule_decay)], alpha=0.8)
        plt.plot(temp_data['Itterations'], temp_data['mean_test_accuracy'], label=f'{schedule_decays_names[schedule_decays.index(schedule_decay)]} (Test)', color=color[schedule_decays.index(schedule_decay)], alpha=0.7, linestyle='--')
        upper_bound = temp_data['mean_train_accuracy'] + (temp_data['std_train_accuracy']/2)
        lower_bound = temp_data['mean_train_accuracy'] - (temp_data['std_train_accuracy']/2)
        plt.fill_between(temp_data['Itterations'], upper_bound, lower_bound, color=color[schedule_decays.index(schedule_decay)], alpha=0.2)

    plt.xlabel('Iterations', fontsize=other_title)
    plt.ylabel('Accuracy', fontsize=other_title)
    full_title = 'Varying Decay Funtion\nFor Simulated Annealing'
    plt.title(full_title, fontsize=header)
    plt.legend(loc='lower right', frameon=True, edgecolor='black', facecolor='white', fontsize=other_title)
    plt.tick_params(direction='in', bottom=True, top=True, left=True, right=True, length=4)
    plt.gca().spines['top'].set_color('black')
    plt.gca().spines['bottom'].set_color('black')
    plt.gca().spines['left'].set_color('black')
    plt.gca().spines['right'].set_color('black')
    spine_alpha = 0.5
    plt.gca().spines['top'].set_alpha(spine_alpha)
    plt.gca().spines['bottom'].set_alpha(spine_alpha)
    plt.gca().spines['left'].set_alpha(spine_alpha)
    plt.gca().spines['right'].set_alpha(spine_alpha)
    plt.grid(True, linestyle="dotted", alpha=0.45, color='black')
    plt.xticks(fontsize=other_title)
    plt.yticks(fontsize=other_title)
    h1 = 'C:/Users/sagni/Documents/Personal Files/CS7641/Assignment2/NN_Images/NN_SA_DecayFn_gridsearch.png'
    plt.savefig(h1, dpi=600)
    plt.show()
    
    return Output
```

```python
def SA_Decay(trainX, trainY, testX, testY):
    
    gc.collect()
    max_iters = [200 * (i+1) for i in range(4)]
    seeds = [1,123,300,999,12345]
    decays = [0.99,0.9,0.8]
    color = ['blue', 'green', 'purple']
    algorithm = 'simulated_annealing'
    header = 24
    other_title = 18
    Output = pd.DataFrame(columns=['algorithm', 'random_seed', 'Itterations', 'decays', 'train_accuracy', 'test_accuracy'])
    
    for max_iter in max_iters:
        for decay in decays:
            for seed in seeds:
                gc.collect()
                # Define neural network classifier
                nn = NeuralNetwork(hidden_nodes=[8,], activation='relu',
                                       algorithm=algorithm, max_iters=int(max_iter),
                                       bias=True, is_classifier=True, learning_rate=0.1,schedule=mlrose.GeomDecay(decay=decay),
                                       early_stopping=False, random_state=seed, curve=True)

                nn.fit(trainX, trainY)
                      
                train_accuracy = accuracy_score(trainY, pd.DataFrame(nn.predict(trainX), index=trainY.index))
                test_accuracy = accuracy_score(testY, pd.DataFrame(nn.predict(testX), index=testY.index))

                Output = Output.append({'algorithm' : algorithm,
                                        'random_seed' : seed,
                                        'Itterations' : max_iter,
                                        'decays' : decay,
                                        'train_accuracy' : train_accuracy,
                                        'test_accuracy' : test_accuracy}, ignore_index=True)
                display(Output)
    
    Output = Output.groupby(['decays', 'Itterations']).agg({'train_accuracy': ['mean', 'std'], 'test_accuracy': ['mean', 'std']}).reset_index()
    # Rename the columns
    Output.columns = ['decays', 'Itterations', 'mean_train_accuracy', 'std_train_accuracy', 'mean_test_accuracy', 'std_test_accuracy']
    
    
    #################################
    ####plot train accuracy
    #################################
    # Plot fitness vs iteration with separate lines for each temperature
    plt.clf()
    plt.figure(figsize=(12, 8))
    plt.gcf().set_facecolor('white')
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.set_facecolor('white')
    ax.set_facecolor('white')
    for decay in decays:
        temp_data = Output[Output['decays'] == decay]
        plt.plot(temp_data['Itterations'], temp_data['mean_train_accuracy'], label=f'Decay (Train) {decay}', color=color[decays.index(decay)], alpha=0.8)
        plt.plot(temp_data['Itterations'], temp_data['mean_test_accuracy'], label=f'Decay (Test) {decay}', color=color[decays.index(decay)], alpha=0.7, linestyle='--')
        upper_bound = temp_data['mean_train_accuracy'] + (temp_data['std_train_accuracy']/4)
        lower_bound = temp_data['mean_train_accuracy'] - (temp_data['std_train_accuracy']/4)
        plt.fill_between(temp_data['Itterations'], upper_bound, lower_bound, color=color[decays.index(decay)], alpha=0.2)

    plt.xlabel('Iterations', fontsize=other_title)
    plt.ylabel('Accuracy', fontsize=other_title)
    full_title = 'Varying Decay Rate\nFor Simulated Annealing'
    plt.title(full_title, fontsize=header)
    plt.legend(loc='lower right', frameon=True, edgecolor='black', facecolor='white', fontsize=other_title)
    plt.tick_params(direction='in', bottom=True, top=True, left=True, right=True, length=4)
    plt.gca().spines['top'].set_color('black')
    plt.gca().spines['bottom'].set_color('black')
    plt.gca().spines['left'].set_color('black')
    plt.gca().spines['right'].set_color('black')
    spine_alpha = 0.5
    plt.gca().spines['top'].set_alpha(spine_alpha)
    plt.gca().spines['bottom'].set_alpha(spine_alpha)
    plt.gca().spines['left'].set_alpha(spine_alpha)
    plt.gca().spines['right'].set_alpha(spine_alpha)
    plt.grid(True, linestyle="dotted", alpha=0.45, color='black')
    plt.xticks(fontsize=other_title)
    plt.yticks(fontsize=other_title)
    h1 = 'C:/Users/sagni/Documents/Personal Files/CS7641/Assignment2/NN_Images/NN_SA_Decay_gridsearch.png'
    plt.savefig(h1, dpi=600)
    plt.show()
    
    return Output
```

```python
#stratified sampling in sklearn
def stratified_sampling(vegito, target, size = 0.4, seed = 903949505):
    gc.collect()
    goku, vegita = train_test_split(vegito, test_size=size, stratify=vegito[target], random_state=seed)
    print(color.BOLD + color.UNDERLINE + color.DARKCYAN + "Original distribution" + color.END)
    c = vegito[target].value_counts(normalize=False)
    p = vegito[target].value_counts(normalize=True)*100
    display(pd.concat([c,p], axis=1, keys=['counts', '%']))
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.WARNING + "Train distribution" + color.END)
    c = goku[target].value_counts(normalize=False)
    p = goku[target].value_counts(normalize=True)*100
    display(pd.concat([c,p], axis=1, keys=['counts', '%']))
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.RED + "Test distribution" + color.END)
    c = vegita[target].value_counts(normalize=False)
    p = vegita[target].value_counts(normalize=True)*100
    display(pd.concat([c,p], axis=1, keys=['counts', '%']))
    
    #split goku, vegita and cell by target variable to make depent and indepent data seperately
    goku_X = goku.drop(target, axis=1)
    vegita_X = vegita.drop(target, axis=1)
    goku_y = pd.DataFrame(goku[target])
    vegita_y = pd.DataFrame(vegita[target])
    
    #return goku_X, goku_y, vegita_X, vegita_y
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_x shape" + color.END)
    print("Number of columns are " + color.BOLD + str(goku_X.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(goku_X.shape[1]) + color.END)
    display(goku_X.head())
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_y shape" + color.END)
    print("Number of columns are " + color.BOLD + str(goku_y.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(goku_y.shape[1]) + color.END)
    display(goku_y.head())
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_x shape" + color.END)
    print("Number of columns are " + color.BOLD + str(vegita_X.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(vegita_X.shape[1]) + color.END)
    display(vegita_X.head())
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_y shape" + color.END)
    print("Number of columns are " + color.BOLD + str(vegita_y.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(vegita_y.shape[1]) + color.END)
    display(vegita_y.head())
    return goku_X, goku_y, vegita_X, vegita_y
```
