```python
import numpy as np
import matplotlib.pyplot as plt
import time
import utils
import pandas as pd
import gc
import mlrose_hiive as mlrose

from mlrose_hiive.algorithms.decay import ExpDecay
from mlrose_hiive.neural import NeuralNetwork

from sklearn.metrics import log_loss, classification_report, accuracy_score, f1_score
from sklearn.model_selection import train_test_split

class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    END = '\033[0m'
```

```python
def NN_GA_mut_prob(trainX, trainY, testX, testY):
    
    gc.collect()
    max_iters = [20 * (i+1) for i in range(10)]
    seeds = [10 * (10**i) for i in range(5)]
    mutation_probs = [0.05, 0.5, 0.8]
    color = ['blue', 'green', 'purple']
    algorithm = 'genetic_alg'
    header = 24
    other_title = 18
    Output = pd.DataFrame(columns=['algorithm', 'random_seed', 'Itterations', 'mutation_probs', 'train_accuracy', 'test_accuracy'])
    
    for max_iter in max_iters:
        for mutation_prob in mutation_probs:
            for seed in seeds:
                gc.collect()
                # Define neural network classifier
                nn = NeuralNetwork(hidden_nodes=[10,], activation='relu',
                                       algorithm=algorithm, max_iters=int(max_iter),
                                       bias=True, is_classifier=True, learning_rate=0.001,
                                       early_stopping=False,
                                       pop_size=50, mutation_prob=mutation_prob,
                                       max_attempts=int(max_iter), random_state=seed, curve=True)

                nn.fit(trainX, trainY)
                      
                train_accuracy = accuracy_score(trainY, pd.DataFrame(nn.predict(trainX), index=trainY.index))
                test_accuracy = accuracy_score(testY, pd.DataFrame(nn.predict(testX), index=testY.index))

                Output = Output.append({'algorithm' : algorithm,
                                        'random_seed' : seed,
                                        'Itterations' : max_iter,
                                        'mutation_probs' : mutation_prob,
                                        'train_accuracy' : train_accuracy,
                                        'test_accuracy' : test_accuracy}, ignore_index=True)
                
                display(Output)
    
    Output['Itterations'] = 200 - Output['Itterations']
    Output = Output.groupby(['mutation_probs', 'Itterations']).agg({'train_accuracy': ['mean', 'std'], 'test_accuracy': ['mean', 'std']}).reset_index()
    # Rename the columns
    Output.columns = ['mutation_probs', 'Itterations', 'mean_train_accuracy', 'std_train_accuracy', 'mean_test_accuracy', 'std_test_accuracy']
    
    
    #################################
    ####plot train accuracy
    #################################
    # Plot fitness vs iteration with separate lines for each temperature
    plt.clf()
    plt.figure(figsize=(12, 8))
    plt.gcf().set_facecolor('white')
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.set_facecolor('white')
    ax.set_facecolor('white')
    for mutation_prob in mutation_probs:
        temp_data = Output[Output['mutation_probs'] == mutation_prob]
        plt.plot(temp_data['Itterations'], temp_data['mean_train_accuracy'], label=f'Mut Prob (Train) {mutation_prob}', color=color[mutation_probs.index(mutation_prob)], alpha=0.8)
        plt.plot(temp_data['Itterations'], temp_data['mean_test_accuracy'], label=f'Mut Prob (Test) {mutation_prob}', color=color[mutation_probs.index(mutation_prob)], alpha=0.7, linestyle='--')
        upper_bound = temp_data['mean_train_accuracy'] + temp_data['std_train_accuracy']
        lower_bound = temp_data['mean_train_accuracy'] - temp_data['std_train_accuracy']
        plt.fill_between(temp_data['Itterations'], upper_bound, lower_bound, color=color[mutation_probs.index(mutation_prob)], alpha=0.2)

    plt.xlabel('Iterations', fontsize=other_title)
    plt.ylabel('Accuracy', fontsize=other_title)
    full_title = 'Varying Mutation Probabilities\nFor Genetic Algorithm'
    plt.title(full_title, fontsize=header)
    plt.legend(loc='lower right', frameon=True, edgecolor='black', facecolor='white', fontsize=other_title)
    plt.tick_params(direction='in', bottom=True, top=True, left=True, right=True, length=4)
    plt.gca().spines['top'].set_color('black')
    plt.gca().spines['bottom'].set_color('black')
    plt.gca().spines['left'].set_color('black')
    plt.gca().spines['right'].set_color('black')
    spine_alpha = 0.5
    plt.gca().spines['top'].set_alpha(spine_alpha)
    plt.gca().spines['bottom'].set_alpha(spine_alpha)
    plt.gca().spines['left'].set_alpha(spine_alpha)
    plt.gca().spines['right'].set_alpha(spine_alpha)
    plt.grid(True, linestyle="dotted", alpha=0.45, color='black')
    plt.xticks(fontsize=other_title)
    plt.yticks(fontsize=other_title)
    h1 = 'C:/Users/sagni/Documents/Personal Files/CS7641/Assignment2/NN_Images/NN_GA_mut_prob_gridsearch.png'
    plt.savefig(h1, dpi=600)
    plt.show()
```

```python
def NN_GA_pop_size(trainX, trainY, testX, testY):
    
    gc.collect()
    max_iters = [50,100,150,200,250]
    seeds = [10 * (10**i) for i in range(5)]
    pop_sizes = [50, 100, 200]
    color = ['blue', 'green', 'purple']
    algorithm = 'genetic_alg'
    header = 24
    other_title = 18
    Output = pd.DataFrame(columns=['algorithm', 'random_seed', 'Iterations', 'pop_sizes', 'train_accuracy', 'test_accuracy'])
    
    for max_iter in max_iters:
        for pop_size in pop_sizes:
            for seed in seeds:
                gc.collect()
                # Define neural network classifier
                nn = NeuralNetwork(hidden_nodes=[10,], activation='relu',
                                       algorithm=algorithm, max_iters=int(max_iter),
                                       bias=True, is_classifier=True, learning_rate=0.001,
                                       early_stopping=False, schedule=ExpDecay(),
                                       pop_size=pop_size, mutation_prob=0.5,
                                       max_attempts=int(max_iter), random_state=seed, curve=True)

                nn.fit(trainX, trainY)
                      
                train_accuracy = accuracy_score(trainY, pd.DataFrame(nn.predict(trainX), index=trainY.index))
                test_accuracy = accuracy_score(testY, pd.DataFrame(nn.predict(testX), index=testY.index))

                Output = Output.append({'algorithm' : algorithm,
                                        'random_seed' : seed,
                                        'Iterations' : max_iter,
                                        'pop_sizes' : pop_size,
                                        'train_accuracy' : train_accuracy,
                                        'test_accuracy' : test_accuracy}, ignore_index=True)
                display(Output)
    
    Output['Iterations'] = 250 - Output['Iterations']
    Output = Output.groupby(['pop_sizes', 'Iterations']).agg({'train_accuracy': ['mean', 'std'], 'test_accuracy': ['mean', 'std']}).reset_index()
    # Rename the columns
    Output.columns = ['pop_sizes', 'Iterations', 'mean_train_accuracy', 'std_train_accuracy', 'mean_test_accuracy', 'std_test_accuracy']
    
    
    #################################
    ####plot train accuracy
    #################################
    # Plot fitness vs iteration with separate lines for each temperature
    plt.clf()
    plt.figure(figsize=(12, 8))
    plt.gcf().set_facecolor('white')
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.set_facecolor('white')
    ax.set_facecolor('white')
    for pop_size in pop_sizes:
        temp_data = Output[Output['pop_sizes'] == pop_size]
        plt.plot(temp_data['Iterations'], temp_data['mean_train_accuracy'], label=f'Pop Size (Train) {pop_size}', color=color[pop_sizes.index(pop_size)], alpha=0.8)
        plt.plot(temp_data['Iterations'], temp_data['mean_test_accuracy'], label=f'Pop Size (Test) {pop_size}', color=color[pop_sizes.index(pop_size)], alpha=0.7, linestyle='--')
        upper_bound = temp_data['mean_train_accuracy'] + temp_data['std_train_accuracy']
        lower_bound = temp_data['mean_train_accuracy'] - temp_data['std_train_accuracy']
        plt.fill_between(temp_data['Iterations'], upper_bound, lower_bound, color=color[pop_sizes.index(pop_size)], alpha=0.2)

    plt.xlabel('Iterations', fontsize=other_title)
    plt.ylabel('Accuracy', fontsize=other_title)
    full_title = 'Varying Population Size\nFor Genetic Algorithm'
    plt.title(full_title, fontsize=header)
    plt.legend(loc='upper left', frameon=True, edgecolor='black', facecolor='white', fontsize=other_title)
    plt.tick_params(direction='in', bottom=True, top=True, left=True, right=True, length=4)
    plt.gca().spines['top'].set_color('black')
    plt.gca().spines['bottom'].set_color('black')
    plt.gca().spines['left'].set_color('black')
    plt.gca().spines['right'].set_color('black')
    spine_alpha = 0.5
    plt.gca().spines['top'].set_alpha(spine_alpha)
    plt.gca().spines['bottom'].set_alpha(spine_alpha)
    plt.gca().spines['left'].set_alpha(spine_alpha)
    plt.gca().spines['right'].set_alpha(spine_alpha)
    plt.grid(True, linestyle="dotted", alpha=0.45, color='black')
    plt.xticks(fontsize=other_title)
    plt.yticks(fontsize=other_title)
    h1 = 'C:/Users/sagni/Documents/Personal Files/CS7641/Assignment2/NN_Images/NN_GA_pop_size_gridsearch.png'
    plt.savefig(h1, dpi=600)
    plt.show()
```

```python
#stratified sampling in sklearn
def stratified_sampling(vegito, target, size = 0.4, seed = 903949505):
    gc.collect()
    goku, vegita = train_test_split(vegito, test_size=size, stratify=vegito[target], random_state=seed)
    print(color.BOLD + color.UNDERLINE + color.DARKCYAN + "Original distribution" + color.END)
    c = vegito[target].value_counts(normalize=False)
    p = vegito[target].value_counts(normalize=True)*100
    display(pd.concat([c,p], axis=1, keys=['counts', '%']))
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.WARNING + "Train distribution" + color.END)
    c = goku[target].value_counts(normalize=False)
    p = goku[target].value_counts(normalize=True)*100
    display(pd.concat([c,p], axis=1, keys=['counts', '%']))
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.RED + "Test distribution" + color.END)
    c = vegita[target].value_counts(normalize=False)
    p = vegita[target].value_counts(normalize=True)*100
    display(pd.concat([c,p], axis=1, keys=['counts', '%']))
    
    #split goku, vegita and cell by target variable to make depent and indepent data seperately
    goku_X = goku.drop(target, axis=1)
    vegita_X = vegita.drop(target, axis=1)
    goku_y = pd.DataFrame(goku[target])
    vegita_y = pd.DataFrame(vegita[target])
    
    #return goku_X, goku_y, vegita_X, vegita_y
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_x shape" + color.END)
    print("Number of columns are " + color.BOLD + str(goku_X.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(goku_X.shape[1]) + color.END)
    display(goku_X.head())
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_y shape" + color.END)
    print("Number of columns are " + color.BOLD + str(goku_y.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(goku_y.shape[1]) + color.END)
    display(goku_y.head())
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_x shape" + color.END)
    print("Number of columns are " + color.BOLD + str(vegita_X.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(vegita_X.shape[1]) + color.END)
    display(vegita_X.head())
    print("\n" + "\n" + color.BOLD + color.UNDERLINE + color.BLUE + "Train_y shape" + color.END)
    print("Number of columns are " + color.BOLD + str(vegita_y.shape[0]) + color.END + " and number of rows are " + color.BOLD + str(vegita_y.shape[1]) + color.END)
    display(vegita_y.head())
    return goku_X, goku_y, vegita_X, vegita_y
```
